```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

## Data Collection

df=pd.read_excel(r"/content/sample_data/heart_disease_data_with_labels.xlsx")
print(df)
```

```python
##Data preprocessing

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df['Chest Pain Type']=le.fit_transform(df['Chest Pain Type'])
df['Fasting Blood Sugar > 120 mg/dl']=le.fit_transform(df['Fasting Blood Sugar > 120 mg/dl'])
df['Resting ECG']=le.fit_transform(df['Resting ECG'])
df['Exercise Induced Angina']=le.fit_transform(df['Exercise Induced Angina'])
df['Slope of Peak Exercise ST Segment']=le.fit_transform(df['Slope of Peak Exercise ST Segment'])
df['Thalassemia (Thal)']=le.fit_transform(df['Thalassemia (Thal)'])

print(df)
```

```python
## Defining input,Output

x=df.drop(columns=['Heart Disease'])
y=df['Heart Disease']
```

```python
## 80% of data for train
## 20% of data for test


from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=12)

print("df",df.shape)
print("x_train",x_train.shape)
print("x_test",x_test.shape)
print("y_train",y_train.shape)
print("y_test",y_test.shape)
```

```python
##Modal Training
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
## Model Evalution

y_predict=model.predict(x_test)
print("Y_predict",y_predict)
print("y_test",y_test)
```

```python
## Accuracy

from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_predict)
print("Accuracy",accuracy)
```

```python
## model Prediction

test_data=model.predict([[36,1,0,109,281,1,2,99,0,2.5,2,1,0,0]])
if test_data==1:
  print("Person has heart disease")
else:
  print("Person does not have heart disease")
```

```python
# prompt: How to increase accuracy

import pandas as pd
import warnings
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
warnings.filterwarnings("ignore")

## Data Collection

df=pd.read_excel(r"/content/sample_data/heart_disease_data_with_labels.xlsx")

##Data preprocessing

le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df['Chest Pain Type']=le.fit_transform(df['Chest Pain Type'])
df['Fasting Blood Sugar > 120 mg/dl']=le.fit_transform(df['Fasting Blood Sugar > 120 mg/dl'])
df['Resting ECG']=le.fit_transform(df['Resting ECG'])
df['Exercise Induced Angina']=le.fit_transform(df['Exercise Induced Angina'])
df['Slope of Peak Exercise ST Segment']=le.fit_transform(df['Slope of Peak Exercise ST Segment'])
df['Thalassemia (Thal)']=le.fit_transform(df['Thalassemia (Thal)'])

## Defining input,Output

x=df.drop(columns=['Heart Disease'])
y=df['Heart Disease']

## Feature Scaling
scaler = StandardScaler()
x = scaler.fit_transform(x)

## 80% of data for train
## 20% of data for test

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

##Modal Training
# Using Random Forest Classifier (often more accurate than Naive Bayes)
model=RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train,y_train)

## Model Evalution

y_predict=model.predict(x_test)

## Accuracy

accuracy=accuracy_score(y_test,y_predict)
print("Accuracy",accuracy)

## model Prediction

test_data=model.predict([[36,1,0,109,281,1,2,99,0,2.5,2,1,0]])
if test_data==1:
  print("Person has heart disease")
else:
  print("Person does not have heart disease")
```
